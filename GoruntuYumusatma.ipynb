```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
resim1= cv2.imread("/content/gdrive/MyDrive/filter.jpg")
```

```python
cekirdek=np.ones((5,5),np.float32)/25  #5x5lik kernelde 25 piksel oldugundan ccekirdegin merkezindeki degeri bulmak icin 25'e boluyoruz.
#ayrica sectigimiz matris sayisi ek sayi olmali cunku cift sayinin merkez noktasi olmaz
resim2=cv2.filter2D(resim1,-1,cekirdek)
```

```python
#iki resmi yan yana inceleyelim
plt.subplot(121), plt.imshow(resim1), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(resim2), plt.title("ortalama filtre")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
resim3= cv2.GaussianBlur(resim1,(5,5),0)
```

```python
plt.subplot(121), plt.imshow(resim1), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(resim3), plt.title("Gaussian filtre")
plt.xticks([]), plt.yticks([])

plt.show()
```
